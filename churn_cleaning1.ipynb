```python
# Convert TotalCharges from string to numeric, forcing errors to NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Now check how many rows became NaN (missing)
df['TotalCharges'].isnull().sum()
```

```python
# Drop the few rows where TotalCharges could not be converted
df.dropna(subset=['TotalCharges'], inplace=True)

# Confirm it again
df['TotalCharges'].isnull().sum()
```

```python
# Confirm type again
df.dtypes
```

```python
# Get all columns that are "object" (i.e., text)
cat_cols = df.select_dtypes(include='object').columns
cat_cols
```

```python
df.drop('customerID', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a label encoder object
le = LabelEncoder()

# Loop through all categorical columns and encode them
for col in cat_cols:
    if col != 'customerID':  # we already dropped it
        df[col] = le.fit_transform(df[col])
```
